```python
# import os, json, numpy as np, pandas as pd

base = "data/bonus"
X = np.load(os.path.join(base, "X.npy"))
T = np.load(os.path.join(base, "T.npy"))
Y = np.load(os.path.join(base, "Y.npy"))
meta = json.load(open(os.path.join(base, "meta.json")))

print("meta x_cols len:", len(meta["x_cols"]))
print("t_col:", meta["t_col"], "| y_col:", meta["y_col"])
print("X shape:", X.shape, "| T:", T.shape, "| Y:", Y.shape)

print("T unique:", np.unique(T))
print("Y dtype:", Y.dtype, "| Y nan count:", np.isnan(Y).sum())

print("T mean:", float(T.mean()))
print("Y mean/std:", float(Y.mean()), float(Y.std()))

# 교집합 점검(원본 df가 있다면)
df = pd.read_csv(os.path.join(base, "bonus_raw.csv"))
print("교집합(T in X_cols):", set([meta["t_col"]]) & set(meta["x_cols"]))
print("교집합(Y in X_cols):", set([meta["y_col"]]) & set(meta["x_cols"]))
```
